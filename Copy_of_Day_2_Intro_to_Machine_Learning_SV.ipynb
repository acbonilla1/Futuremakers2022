# Machine Learning

Machine learning is a subfield of artificial intelligence (AI) that enables systems to automatically learn and improve from examples and experience (aka data), without being explicitly programmed.

## Supervised Learning

Machines often learn from sample data that has both an example input and an example output. For example, a data input-output sample pair may be input data about an individual’s credit history, and their corresponding credit risk (either specified by a human expert or based on historical outcomes). Given enough of these input-output samples, the machine learns how to construct a “model“ (i.e., a logic or algorithm) that represents the relationship between an individual's credit history and how 'risky' they might be considered from a credit perspective, based on the samples it has seen. 

This class of machine learning is called “supervised learning”, since along with the input data, the desired outcome is also given, and the model is “supervised” or guided by the outcome in learning the associated model parameters. 

There are 2 types of supervised learning approaches: a) **Regression** , b) **Classification** [Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

## Unsupervised Learning

Another form of machine learning is "unsupervised learning". In unsupervised learning, the machine learns from data for which the outcomes are not known. It’s given input samples, but no output samples.

There are 2 types of unsupervised learning approaches: a) **Clustering**, b) **Dimensionality Reduction** [Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

image_2021-10-31_201022.png

##Semi-supervised Learning

Supervised and unsupervised learning each have their pros and cons. Supervised learning is typically demanding from the standpoint of acquiring enough input-output labeled data for learning, while unsupervised learning is relatively less demanding in that regard because it only needs input data but its outcomes can be unpredictable. 

A middle ground is semi-supervised learning. In semi-supervised learning, a model uses the larger  unlabeled data to gain a general sense of the data’s structure, and a much smaller subset of labeled data to learn how to group and organize the whole dataset. [Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

## Reinforcement Learning

Reinforcement learning is like unsupervised learning in the sense that outputs are usually not given.

The central concept of reinforcement learning is based around an “agent” (a computer or robot) that is interacting with an “environment” (here defined as everything that is not the agent).

The agent performs actions on the environment (for instance, a robot takes a step forward). Then, the environments will then provide some sort of feedback to the agent, usually in a form called the “reward.” [Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

While supervised learning is about instructive feedback, reinforcement learning is about **evaluative feedback**.

Iris data sets consists of 3 different types of irises’ (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray

The rows represent the samples. The columns represent the attribtes (also called features) of the samples: Sepal Length, Sepal Width, Petal Length and Petal Width.

```python
# Load in our dataset
iris_data = load_iris()
```

Let's visualize the data first

```python
import matplotlib.pyplot as plt
```

```python
# we only take the first two features.
X = iris_data.data[:, :2]  
# we get all the target for each sample
y = iris_data.target       

# we get minimum and maximum values so that we know what should be plot axis ticks 
# (this is only for visualization purpose)
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5  
# we get minimum and maximum values so that we know what should be plot axis ticks 
# (this is only for visualization purpose)
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5  

# we will be creating 2 figures, each of size 8 x 6 dpi
plt.figure(2, figsize=(8, 6))   
# clear any previous figures (if any), and make this the current figure
plt.clf()                       

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.title('Red = Setosa, Yellow = Versicolour, Gray = Virginica')

plt.show()
```

# Decision Trees

Decision trees for both classification and regression are super easy to use in Scikit Learn with a built in class.

We’ll initialise our decision tree for classification, which is also a built in class.

```python
# Initialize our decision tree object
classification_tree = tree.DecisionTreeClassifier()
```

Running training is then a simple one-liner! The $.fit(X, Y)$ function trains the model where X is the NumPy array of inputs and Y is the corresponding NumPy array of outputs.

```python
# Train our decision tree (tree induction + pruning)
classification_tree = classification_tree.fit(iris_data.data, iris_data.target)
```

# Graphviz

You have been acquainted with **Matplotlib**, now let's use a different visualization library in Python: **Graphviz**

```python
import graphviz 
```

Graphviz comes with a few options that will help in visualising the decision nodes and splits that the model learned which is super useful for understanding how it all works. 

```python
dot_data = tree.export_graphviz(classification_tree, out_file=None, feature_names=iris_data.feature_names, class_names=iris_data.target_names, filled=True, rounded=True, special_characters=True)  
```

Render the decision tree in a .pdf file

```python
graph = graphviz.Source(dot_data)  
graph.render("iris") 
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

iris_data = load_iris()
train_data, test_data, train_target, test_target = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=0)
print(train_data.shape, test_data.shape, train_target.shape, test_target.shape)
total_test_samples = test_target.shape[0]

print("----------------------------------------------------------")
print("Decision Tree Performance")
classification_tree = tree.DecisionTreeClassifier()
classification_tree = classification_tree.fit(train_data, train_target)
classification_tree_predictions = classification_tree.predict(test_data)
print("Predictions: ", classification_tree_predictions)
print("Ground Truth:", test_target)

classification_tree_correct = sum(classification_tree_predictions == test_target)
classification_tree_accuracy = classification_tree_correct / total_test_samples * 100
print("Decision Tree Accuracy:", classification_tree_accuracy, "%")
print("----------------------------------------------------------")

print("Random Forest Performance")
classification_forest = RandomForestClassifier(n_estimators=10)
classification_forest = classification_forest.fit(train_data, train_target)
classification_forest_predictions = classification_forest.predict(test_data)
print("Predictions: ", classification_forest_predictions)
print("Ground Truth:", test_target)

classification_forest_correct = sum(classification_forest_predictions == test_target)
classification_forest_accuracy = classification_forest_correct / total_test_samples * 100
print("Random Forest Accuracy:", classification_forest_accuracy, "%")
```
